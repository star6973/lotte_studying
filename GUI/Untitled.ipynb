```python
import sys
import cv2
import PyQt5
from PyQt5.QtCore import QTimer
from PyQt5.QtGui import QImage
from PyQt5.QtGui import QPixmap
from PyQt5 import QtGui, QtCore
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *
from PyQt5.QtGui import *
from PyQt5.QtWidgets import QApplication
from PyQt5.QtWidgets import QDialog
from PyQt5.uic import loadUi
import sys
from functools import partial
import numpy as np

class ControlWindow(QDialog):
    def __init__(self):
        super(ControlWindow, self).__init__()
        loadUi('third(info).ui', self)
        
        self.capture = cv2.VideoCapture(-1)
        self.capture.set(cv2.CAP_PROP_FRAME_HEIGHT,480)
        self.capture.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
        
        self.timer = QTimer(self)
        self.timer.timeout.connect(self.update_frame)
        self.timer.start(3)
        
        self.setWindowTitle('Control Panel')
        self.show()

    def update_frame(self):
        _, self.image = self.capture.read()
        self.image = cv2.flip(self.image, 1)
        self.displayImage(self.image)
        
        self.front_capture_button.clicked.connect(self.capture_front)
#         self.left_capture_button.clicked.connect(self.capture_and_save_image)
#         self.right_capture_button.clicked.connect(self.capture_and_save_image)
#         self.up_capture_button.clicked.connect(self.capture_and_save_image)
#         self.down_capture_button.clicked.connect(self.capture_and_save_image)
        
    def capture_front(self):
        _, frame = self.capture.read()
        qformat = QImage.Format_Indexed8
        if len(frame.shape) == 3:
            if frame.shape[2] == 4:
                qformat = QImage.Format_RGBA8888
            else:
                qformat = QImage.Format_RGB888

        outImage = QImage(frame, frame.shape[1], frame.shape[0], frame.strides[0], qformat)
        outImage = outImage.rgbSwapped()
        outImage.save('front_img.jpg')
        
        from PIL.ImageQt import ImageQt
        from PIL import Image
        
        image2 = Image.open('front_img.jpg')
        qimage = ImageQt(image2)
        pixmap = QtGui.QPixmap.fromImage(qimage)
        
        
        self.front_label.setPixmap(pixmap)
        self.front_label.setScaledContents(True)
        
#     def capture_and_save_image(self):
# #         _, frame = self.capture.read()

#         print(num)

#         frame = self.image
        
#         # 정면 얼굴 사진
#         if num == 0:
#             cv2.imwrite(filename='front_img.jpg', img=frame)
#             front_image = cv2.imread('front_img.jpg', 1)
#             self.front_label.setPixmap(QPixmap.fromImage(front_image))
#             self.front_label.setScaledContents(True)
        
#         # 왼쪽 얼굴 사진
#         elif num == 1:
#             cv2.imwrite(filename='left_img.jpg', img=frame)
#             left_image = cv2.IMREAD_COLOR('left_img.jpg', 1)
#             self.left_label.setPixmap(QPixmap.fromImage(left_image))
#             self.left_label.setScaledContents(True)
        
#         # 오른쪽 얼굴 사진
#         elif num == 2:
#             cv2.imwrite(filename='right_img.jpg', img=frame)
#             right_image = cv2.IMREAD_COLOR('right_img.jpg', 1)
#             self.right_label.setPixmap(QPixmap.fromImage(right_image))
#             self.right_label.setScaledContents(True)
            
#         # 위쪽 얼굴 사진
#         elif num == 3:
#             cv2.imwrite(filename='up_img.jpg', img=frame)
#             up_image = cv2.IMREAD_COLOR('up_img.jpg', 1)
#             self.up_label.setPixmap(QPixmap.fromImage(up_image))
#             self.up_label.setScaledContents(True)
        
#         # 아래쪽 얼굴 사진
#         elif num == 4:
#             cv2.imwrite(filename='down_img.jpg', img=frame)
#             down_image = cv2.IMREAD_COLOR('down_img.jpg', 1)
#             self.down_label.setPixmap(QPixmap.fromImage(down_image))
#             self.down_label.setScaledContents(True)

    def displayImage(self, img):
        qformat = QImage.Format_Indexed8
        if len(img.shape) == 3:
            if img.shape[2] == 4:
                qformat = QImage.Format_RGBA8888
            else:
                qformat = QImage.Format_RGB888

        outImage = QImage(img,img.shape[1],img.shape[0],img.strides[0],qformat)
        outImage = outImage.rgbSwapped()

        self.video_frame.setPixmap(QPixmap.fromImage(outImage))
        self.video_frame.setScaledContents(True)

if __name__ == '__main__':
    app = QApplication(sys.argv)
    window = ControlWindow()
    window.show()
    app.exec_()
```
