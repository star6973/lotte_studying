```python
from natsort import natsorted

file_dir = os.listdir('./Annotations/')
file_dir = natsorted(file_dir)
file_dir = file_dir[:-1]
print(file_dir)
```

```python
import os

for file in file_dir:
    tree = elemTree.parse(os.path.join('./Annotations', file))
    f = open('./labels/' + file.split('.')[0] + '.txt', 'w')
    for obj in tree.findall('object'):
        class_name = obj.find('name').text
        f.write(class_name + ' ')
        for bnd in obj.find('bndbox'):
            f.write(str(format(float(bnd.text), '.5f')) + ' ')
        f.write('\n')
    f.close()
```

```python
f = open('train.txt', 'r')
lines = f.readlines()
# print(lines)
fr = open('./training/train.txt', 'w')
for line in lines:
    fr.write(line.split('.')[0])
    fr.write('\n')
f.close()
fr.close()
```
