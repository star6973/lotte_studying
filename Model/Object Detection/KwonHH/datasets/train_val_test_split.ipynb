```python
import os
from natsort import natsorted

JPEG_IMAGES = os.listdir("/home/jmh/다운로드/Lotte_XML_new/JPEGImages/")
JPEG_IMAGES = natsorted(JPEG_IMAGES)
print(JPEG_IMAGES)
```

```python
ANNOTATIONS = os.listdir("/home/jmh/다운로드/Lotte_XML_new/Annotations/")
ANNOTATIONS = natsorted(ANNOTATIONS)
print(ANNOTATIONS)
```

```python
# train, val, test = 0.6, 0.2, 0.2 비율로 나누기
import numpy as np
import pandas as pd

df = pd.DataFrame(JPEG_IMAGES, columns=['file_id'])
df
```

```python
train, val, test = np.split(df.sample(frac=1), [int(.6*len(df)), int(.8*len(df))])
```

```python
train = train.sort_index()
train.reset_index(drop=True)
train
```

```python
val = val.sort_index()
val.reset_index(drop=True)
val
```

```python
test = test.sort_index()
test.reset_index(drop=True)
test
```

```python
# 각 데이터프레임에 있는 file_id와 일치하도록, JPEG와 Annoation을 나누기
import shutil

src_img_dir = "/home/jmh/다운로드/Lotte_XML_new/JPEGImages/"
src_ann_dir = "/home/jmh/다운로드/Lotte_XML_new/Annotations/"

dst_img_dir = "/home/jmh/content/datasets/train/images/"
dst_ann_dir = "/home/jmh/content/datasets/train/annotations/"

for tr in train.file_id:
    img_name = tr
    ann_name = tr.split(".")[0] + ".xml"
    shutil.copy(os.path.join(src_img_dir, img_name), os.path.join(dst_img_dir, img_name))
    shutil.copy(os.path.join(src_ann_dir, ann_name), os.path.join(dst_ann_dir, ann_name))
```

```python
dst_img_dir = "/home/jmh/content/datasets/val/images/"
dst_ann_dir = "/home/jmh/content/datasets/val/annotations/"

for vl in val.file_id:
    img_name = vl
    ann_name = vl.split(".")[0] + ".xml"
    shutil.copy(os.path.join(src_img_dir, img_name), os.path.join(dst_img_dir, img_name))
    shutil.copy(os.path.join(src_ann_dir, ann_name), os.path.join(dst_ann_dir, ann_name))
```

```python
dst_img_dir = "/home/jmh/content/datasets/test/images/"
dst_ann_dir = "/home/jmh/content/datasets/test/annotations/"

for ts in test.file_id:
    img_name = ts
    ann_name = ts.split(".")[0] + ".xml"
    shutil.copy(os.path.join(src_img_dir, img_name), os.path.join(dst_img_dir, img_name))
    shutil.copy(os.path.join(src_ann_dir, ann_name), os.path.join(dst_ann_dir, ann_name))
```

```python
# 나누어진 train, val, test 폴더의 annotation 리스트를 텍스트 파일로 만들기
f_tr = open("/home/jmh/content/datasets/train/tr_ann_list.txt", "w")
f_vl = open("/home/jmh/content/datasets/val/vl_ann_list.txt", "w")
f_ts = open("/home/jmh/content/datasets/test/ts_ann_list.txt", "w")

for tr in train.file_id:
    ann_name = tr.split(".")[0] + ".xml"
    f_tr.write(ann_name)
    f_tr.write("\n")
f_tr.close()

for vl in val.file_id:
    ann_name = vl.split(".")[0] + ".xml"
    f_vl.write(ann_name)
    f_vl.write("\n")
f_vl.close()

for ts in test.file_id:
    ann_name = ts.split(".")[0] + ".xml"
    f_ts.write(ann_name)
    f_ts.write("\n")
f_ts.close()
```
